#### Plotting a given line 

```python
# fig = plt.figure()
# fig.add_subplot(111)
x = [1,2,3]
y = [5,7,4]

x2 = [1,2,3]
y2 = [10,14,12]

plt.plot(x,y , label = 'First Line')
plt.plot(x2,y2 , label = 'Second Line')
plt.xlabel('this is the x axis')
plt.ylabel('this is the y axis')
plt.legend()
#plt.xlim(1,3)
#plt.ylim(1,3)
plt.title('This graph is just awesome')
#plt.savefig('pulkit')
plt.show()
```

## plotting a bar chart and the histo grams

```python
plt.bar([1,3,5,7,9],[5,2,7,8,2], label =  'Bar graph1')
plt.bar([2,4,6,8,10],[5,2,7,8,2], label =  'Bar graph2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('This is the most basic.\n Bar Graph')
plt.legend()
```

## Example of the stack plot here

```python
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating =   [2,3,4,3,2]
working =  [7,8,7,2,2]
playing =  [8,5,7,8,13]

plt.stackplot(days,sleeping,eating,working,playing)
plt.show()
```

## Here comes the pie Charts

```python
slices = [12,14, 23, 34]
activities = ['Slepping','racing','proggramming','playing']
cols =  ['c', 'r', 'b', 'm']
plt.pie(slices, labels= activities,colors=cols, startangle=90, shadow=True, explode=(0.1,0.2,0,0))
plt.title('This is the pie chart')
plt.show()
```

```python
x  = []
y  = []
with open('data.csv', 'r') as csvfile:
    plot =  csv.reader(csvfile, delimiter = ',')
    for rows in plot:
        if len(rows) == 0:
            break;
        else:
            x.append(rows[0])
            y.append(rows[1])
plt.plot(x,y)
```

```python
# init  a figure object for plotting 
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(x,y)
```
